```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn import datasets
from sklearn import svm
```

```python
cancer = datasets.load_breast_cancer()
```

```python
print(cancer)
```

```python
data_frame = pd.DataFrame(cancer.data, columns = cancer.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label'] = cancer.target
```

```python
data_frame.shape
data_frame.info()
```

```python
data_frame.isnull().sum()
```

```python
data_frame.describe()
```

```python
data_frame['label'].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
from sklearn.model_selection import train_test_split
```

```python
A = data_frame.drop(columns='label', axis=1)
B = data_frame['label']

print(A)
print(B)
```

```python
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.15, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

A_train_std = scaler.fit_transform(A_train)

A_test_std = scaler.transform(A_test)
```

```python
print(A.shape, A_train.shape, A_test.shape)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras


model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(A_train_std, B_train, validation_split=0.1, epochs=15)
```

```python
plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='lower right')
plt.show()


plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper right')
plt.show()
```

```python
loss, accuracy = model.evaluate(A_test_std, B_test)
print(accuracy)
B_pred = model.predict(A_test_std)
```
